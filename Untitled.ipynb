```python
import numpy as np
import argparse
import dlib
import cv2
import matplotlib.pyplot as plt
import os
import glob
import torch
import sys

from torch import Tensor
from PIL import Image
from torchvision import transforms
from imutils import face_utils
```

```python
def align_face(image, detector, predictor):
    faces = []
    faces.append(np.expand_dims(image, 0))
    faces = np.asarray(faces)

    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    rects = detector(gray, 1)
    
    for (i, rect) in enumerate(rects):
        shape = predictor(gray, rect)
        shape = face_utils.shape_to_np(shape)

        center_eyes = shape[27].astype(np.int)
        eyes_d = np.linalg.norm(shape[36] - shape[45])
        face_size_x = int(eyes_d * 2)

        d = (shape[45] - shape[36]) / eyes_d
        a = np.rad2deg(np.arctan2(d[1], d[0]))
        scale_factor = float(256) / float(face_size_x * 1.4)
        
        M = np.append(cv2.getRotationMatrix2D((center_eyes[0], center_eyes[1]),a,scale_factor),[[0,0,1]], axis=0)
        M1 = np.array([[1, 0, -center_eyes[0] + 128.], 
            [0, 1, -center_eyes[1] + 128.], [0, 0, 1]])
        M = M1.dot(M)[:2]

        try:
            face = cv2.warpAffine(image, M, (256, 256), borderMode=cv2.BORDER_REPLICATE)
            faces.append(face)
        except:
            continue

    return face

def ceneyi_bul(image_input, predictor_input):
    try:
        detector = dlib.get_frontal_face_detector()
        predictor = dlib.shape_predictor(predictor_input)
        image = plt.imread(image_input)
        face = align_face(image, detector, predictor)

        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        yuz = detector(gray, 1)

        shape = predictor(gray, yuz[0])
        shape = face_utils.shape_to_np(shape)

        cene = image[shape[2][1] : shape[8][1], shape[2][0] : shape[14][0]].copy()
        cv2.imwrite("cene_goruntusu.jpg", cv2.cvtColor(cene, cv2.COLOR_RGB2BGR))
    except:
        print("Yuz tespit edilemedi")
    return (face, cene, shape)

def modele_sok(img_input, model_input):
    img = cv2.imread(img_input)
    img = cv2.resize(img, (64, 40))
    img = cv2.copyMakeBorder(cv2.cvtColor(img, cv2.COLOR_RGB2GRAY), 24, 24, 12, 12, cv2.BORDER_CONSTANT)
    img = transforms.ToPILImage(mode='L')(img)
    
    transform = transforms.Compose([        
        transforms.Resize(88),                    
        transforms.ToTensor(),                     
    ])
    
    img = transform(img)
    img = img.unsqueeze(0)

    model = torch.load(model_input).eval()
    model.cpu()

    out = model(img)
    _, indices = torch.sort(out, descending=True)
    
    return indices[0]

def ceneyi_sil(image_input, yuz_noktalari):
    image = plt.imread(image_input)
    image2 = image.copy()
    cene = yuz_noktalari[2:15]
    agiz = yuz_noktalari[48:59]
    
    # Cenesiz goruntu maskesi
    cv2.drawContours(image, [cene], -1, (255, 255, 255), -1, cv2.LINE_AA)
    
    # Agizsiz goruntu maskesinin degili
    mask_agiz = np.zeros(image2.shape[:2], np.uint8)
    cv2.drawContours(mask_agiz, [agiz], -1, (255, 255, 255), -1, cv2.LINE_AA)
    cv2.bitwise_not(mask_agiz, mask_agiz)

    return image
    
def agzi_sil(image_input, yuz_noktalari):
    image = plt.imread(image_input)

    cene = yuz_noktalari[2:15]
    agiz = yuz_noktalari[48:59]
    
    mask_cene = np.zeros(image.shape[:2], np.uint8)
    cv2.drawContours(image, [cene], -1, (255, 255, 255), -1, cv2.LINE_AA)
    mask_agiz = np.zeros(image.shape[:2], np.uint8)
    cv2.drawContours(mask_agiz, [agiz], -1, (255, 255, 255), -1, cv2.LINE_AA)
    
    a = cv2.bitwise_and(image, image, mask=mask_cene)
    a = cv2.bitwise_and(a, a, mask=mask_agiz)

    sakalsiz_goruntu = image - a 

    return sakalsiz_goruntu

```

```python
sys.path.insert(0, "./PyTorchImageClassifier")

path = "face_align/newvasif2.jpg"

# Resmi oku, yuzu sec, yuz noktalarini ve ceneyi tespit et
yuz, cene, yuz_noktalari = ceneyi_bul(path, "face_align/shape_predictor_68_face_landmarks.dat")

# Cenede sakal olup olmadigina karar ver
# 0 : Sakal var, 1 : Sakal yok
sonuc = modele_sok("cene_goruntusu.jpg", "PyTorchImageClassifier/models/conv2_layer_28x28.pt")

# Sakal varsa agzi baska bir yerde tutup yuzu degistir
if(sonuc[0] == 0):
    output = agzi_sil(path, yuz_noktalari)
    cv2.imwrite("yuz.jpg", cv2.cvtColor(yuz, cv2.COLOR_BGR2RGB))
    cv2.imwrite("deneme.jpg", cv2.cvtColor(output, cv2.COLOR_BGR2RGB))
    plt.imshow(output)
else:
    print("Sakal yok")
```

```python
def ceneyi_at(image_input, agiz_input):
    image = plt.imread(image_input)
    agiz_input = cv2.cvtColor(agiz_input, cv2.COLOR_BGR2RGB)
    cene = yuz_noktalari[2:15]
    
    mask = np.zeros(image.shape[:2], np.uint8)
    cv2.drawContours(mask, [cene], -1, (255, 255, 255), -1, cv2.LINE_AA)
    dst = cv2.bitwise_and(image, agiz_input, mask=mask)
    bg = np.ones_like(image, np.uint8)*255
    cv2.bitwise_not(bg, bg, mask=mask)
    agiz = bg + dst
    
    plt.imshow(agiz)
    return agiz
```
